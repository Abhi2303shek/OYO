```python
import pandas as pd
```

```python
oyo = pd.read_csv('OYO_HOTEL_ROOMS.csv', sep=',')
oyo
```

```python
import re

def normalize(text):
    text = str(text).lower()
    text = text.replace('-', ' ')
    text = re.sub(r'\bii\b', '2', text)  # roman II → 2
    text = re.sub(r'\biii\b', '3', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

oyo['hotel'] = oyo['Hotel_name'].apply(normalize)
oyo['address'] = oyo['Location'].apply(normalize)
```

```python
def remove_location(row):
    pattern = r'\b' + re.escape(row['address']) + r'\b'
    return re.sub(pattern, '', row['hotel']).strip()

oyo['hotel'] = oyo.apply(remove_location, axis=1)
```

```python
oyo['hotel'] = oyo['hotel'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
oyo['hotel'] = oyo['hotel'].str.title()
```

```python
oyo['address'] = oyo['address'].str.title()
```

```python
oyo
```

```python
oyo = oyo.drop(columns=['Hotel_name','address'])
```

```python
oyo
```

```python
oyo['Discount'] = oyo['Discount'].str.replace('%', '', regex=False)
```

```python
oyo.columns = ['id','location','price','discount','rating','hotel']
```

```python
oyo = oyo[['id','hotel','location','price','discount','rating']]
```

```python
oyo = oyo.drop_duplicates()
```
